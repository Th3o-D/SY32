## Exercice 1 : Synthèse d’image en Python et NumPy

### Q1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = 70
d = 45

ref = (128, 128)

c = {
    "r": (d * np.cos(np.pi / 6), d * np.sin(np.pi / 6)), 
    "g": (-d * np.cos(np.pi / 6), d * np.sin(np.pi / 6)),
    "b": (0, -d)
}

print(c)
c = {k:(v[0] + 128, 128 - v[1]) for k, v in c.items()}
print(c)
```

### Q2

```python
D = np.zeros((256,256),dtype='float32')
M = np.zeros((256,256),dtype='uint8')

for i in range(256):
    for j in range(256):
        D[i,j] = np.sqrt( (j - c['r'][0])**2 + (i - c['r'][1])**2 )

M = D <= 70
D
```

```python
fig, ax = plt.subplots(1,2)
ax[0].imshow(D)
ax[0].set_title('D')
ax[1].imshow(M, cmap='gray')
ax[1].set_title('M')
plt.show()
```

```python
D = np.zeros((256,256),dtype='float32')
M = np.zeros((256,256),dtype='uint8')
x = np.arange(256)
y = np.arange(256)
xx, yy = np.meshgrid(x, y)
D = np.sqrt( (xx-c['r'][0])**2 + (yy-c['r'][1])**2 )
M = D <= 70
```

### Q4 & Q5

```python
mask = dict()

for color in ('r', 'g', 'b'):
    D = np.zeros((256,256),dtype='float32')
    M = np.zeros((256,256),dtype='uint8')
    x = np.arange(256)
    y = np.arange(256)
    xx, yy = np.meshgrid(x, y)
    D = np.sqrt( (xx-c[color][0])**2 + (yy-c[color][1])**2 )
    M = D <= 70
    mask[color] = M.copy()
    fig, ax = plt.subplots(1,2)
    ax[0].imshow(D)
    ax[0].set_title('D')
    ax[1].imshow(mask[color], cmap='gray')
    ax[1].set_title('M')
    plt.show()

I = np.dstack((mask['r'], mask['g'], mask['b']))
I = I*255

plt.imshow(I)
    
```
