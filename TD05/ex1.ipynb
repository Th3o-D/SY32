```python
import numpy as np
from skimage import data
from utils import *
```

### Q1

```python
camera = data.camera()

Tx, Ty = 10, 20

H = np.array([
    [1, 0, Tx],
    [0, 1, Ty],
    [0, 0, 1],
])

camera_translated = transformer(camera, H)

two_plots('Original', camera, 'Translated', camera_translated)
```

### Q2

```python
r = 0.4
x0 = 100
y0 = 100

H1 = np.array([
    [1, 0, x0],
    [0, 1, y0],
    [0, 0, 1],
])

H2 = np.array([
    [r, 0, 0],
    [0, r, 0],
    [0, 0, 1],
])

H3 = np.array([
    [1, 0, -x0],
    [0, 1, -y0],
    [0, 0, 1],
])

H = H1 @ H2 @ H3

camera_scaled = transformer(camera, H)

two_plots('Original', camera, 'Scaled', camera_scaled)
```

### Q3

```python
alpha = np.deg2rad(20)
cosa = np.cos(alpha)
sina = np.sin(alpha)

H2 = np.array([
    [cosa, sina, 0],
    [-sina, cosa, 0],
    [0, 0, 1],
])

H = H1 @ H2 @ H3

camera_rotated = transformer(camera, H)

two_plots('Original', camera, 'Rotated', camera_rotated)
```
